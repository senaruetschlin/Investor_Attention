```python
# List of all 30 companies in the Dow Jones Industrial Average Index (excluding DOW, since it exists only since 2019)
dow_list = ['AXP', 'AMGN', 'AAPL', 'BA', 'CAT', 'CSCO', 'CVX', 'GS', 'HD', 'HON',
            'IBM', 'INTC', 'JNJ', 'KO', 'JPM', 'MCD', 'MMM', 'MRK', 'MSFT', 'NKE',
            'PG', 'TRV', 'UNH', 'VZ', 'V', 'WBA', 'WMT', 'DIS']

# Dictionary to hold data for each company
company_dataframes = {}

# Loop that fetches the stock data for each company in the list and stores it in the dictionary
for company in dow_list:
    stock_data = yf.download(company, start='2017-12-31', end='2023-01-31')
    
    # Store the data in the dictionary
    company_dataframes[company] = stock_data
```

```python
# see the first row in each dataframe
for company, df in company_dataframes.items():
    print(f'First row of {company}:\n{df.iloc[0]}\n')
```

```python
combined_df = pd.concat(company_dataframes.values(), keys=company_dataframes.keys())
```

```python
# compute the daily return for each company
combined_df['Return'] = (combined_df['Close'] - combined_df['Open']) / combined_df['Open']

# compute the volatility per day for each company
combined_df['Volatility'] = (combined_df['High'] - combined_df['Low']) / combined_df['Open']

# Print the first 5 rows of the DataFrame
print(combined_df.head())
```
